```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

creat the CSV dataset thanks to GPT4o

```python
data = {
    'outlook': ['sunny', 'sunny', 'overcast', 'rainy', 'rainy', 'rainy', 'overcast', 'sunny', 'sunny', 'rainy', 'sunny', 'overcast', 'overcast', 'rainy'],
    'temperature': [85, 80, 83, 70, 68, 65, 64, 72, 69, 75, 75, 72, 81, 71],
    'humidity': [85, 90, 86, 96, 80, 70, 65, 95, 70, 80, 70, 90, 75, 91],
    'windy': [False, True, False, False, False, True, True, False, False, False, True, True, False, True],
    'play': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
}

df = pd.DataFrame(data)
df
```

```python
label_encoder_outlook = LabelEncoder()
df['outlook'] = label_encoder_outlook.fit_transform(df['outlook'])

label_encoder_windy = LabelEncoder()
df['windy'] = label_encoder_windy.fit_transform(df['windy'])

label_encoder_play = LabelEncoder()
df['play'] = label_encoder_play.fit_transform(df['play'])
```

split data

```python
X = df[['outlook', 'temperature', 'humidity', 'windy']]
y = df['play']
```

normaliz

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

split data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

create the NN with relu and sigmoid activation func

```python
model = Sequential()
model.add(Dense(8, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=5, verbose=1)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Accuracy: {accuracy:.2f}')
```
